```python
import pandas as pd
```

```python
df1 = pd.read_csv('Adidas Sales.csv')
```

```python
df1
```

```python
df1.info()
```

```python
df1.drop(columns =['Total Sales'],inplace=True)
```

```python
df1['Price per Unit'] = pd.to_numeric(df1['Price per Unit'].replace('[\$,]', '', regex=True), errors='coerce').astype('Int64')
```

```python
df1
```

```python
df1['Units Sold'] = df1['Units Sold'].replace(',', '', regex=True).astype(float).astype(int)
```

```python
df1['total_sales'] = df1['Price per Unit'] *df1['Units Sold']
```

```python
df1
```

```python
df1['Operating Profit'] = df1['Operating Profit'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df1
```

```python
df1['Region'].unique()
```

```python
total_revenue  = df1['total_sales'].sum()
```

```python
total_revenue
```

```python
units_sold = df1['Units Sold'].sum()
```

```python
average_order_value = total_revenue/units_sold
```

```python
average_order_value
```

```python

import matplotlib.pyplot as plt

region_sales = df1.groupby('Region')['total_sales'].sum().sort_values(ascending=False)


plt.figure(figsize=(10,6))
region_sales.plot(kind='bar', color='seagreen')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df1['Invoice Date'] = pd.to_datetime(df1['Invoice Date'])
df1['YearMonth'] = df1['Invoice Date'].dt.to_period('M').astype(str)
```

```python
df1['Total Sales'] = df1['total_sales'].replace('[\$,]', '', regex=True).astype(float)

monthly_sales = df1.groupby('YearMonth')['Total Sales'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(monthly_sales['YearMonth'], monthly_sales['Total Sales'], 
         marker='o', linestyle='-', color='green', linewidth=2)

plt.title('Monthly Sales Trend', fontsize=16, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Total Sales (USD)')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
category_sales = df1.groupby('Product')['total_sales'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt

# Set the style
plt.figure(figsize=(8, 8))
colors = plt.cm.Greens(range(100, 100 + len(category_sales)*20, 20))  # green shades

plt.pie(category_sales['total_sales'], 
        labels=category_sales['Product'], 
        autopct='%1.1f%%', 
        startangle=140, 
        colors=colors,
        wedgeprops={'edgecolor': 'white'})

plt.title('Product Category Share in Total Revenue', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
